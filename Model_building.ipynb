```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('EDA_data.csv')
```

## Steps to go about:
#### 1. choose relevant columns
#### 2. get dummy data
#### 3. train test split
#### 4. multiple linear regression
#### 5. lasso regression
#### 6. random forest
#### 7. tune models GridsearchCV
#### 8. test ensembles

# 1.

```python
df.columns
```

```python
df_model = df[['Rating', 'Size', 'Type of ownership', 'avg_salary', 'Industry', 'Sector', 'Revenue', 'num_competitors', 'hourly', 'employer_provided', 'job_state'
            , 'same_state', 'age', 'python_yn', 'spark', 'aws', 'excel', 'job_simpl', 'seniority', 'desc_length']]
```

# 2.

```python
df_dum = pd.get_dummies(df_model)
```

# 3.

```python
from sklearn.model_selection import train_test_split

X = df_dum.drop('avg_salary', axis=1)
y = df_dum.avg_salary.values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

# 4.

```python
import statsmodels.api as sm

X_sm = X = sm.add_constant(X)
model = sm.OLS(y, X_sm)
model.fit().summary()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

lm = LinearRegression()
lm.fit(X_train, y_train)

cross_val_score(lm, X_train, y_train, scoring = 'neg_mean_absolute_error', cv=3)
```

```python
np.mean(cross_val_score(lm, X_train, y_train, scoring = 'neg_mean_absolute_error', cv=3))
```

# 5.

```python
from sklearn.linear_model import Lasso

lm_l = Lasso(alpha=.13)
lm_l.fit(X_train, y_train)
np.mean(cross_val_score(lm_l, X_train, y_train, scoring = 'neg_mean_absolute_error', cv=3))
```

```python
alpha = []
error =[]

for i in range(1,100):
    alpha.append(i/100)
    lml = Lasso(alpha=(i/100))
    error.append(np.mean(cross_val_score(lml, X_train, y_train, scoring = 'neg_mean_absolute_error', cv=3)))

plt.plot(alpha, error)
```

```python
err = tuple(zip(alpha,error))
df_err = pd.DataFrame(err, columns=['alpha', 'error'])
df_err[df_err.error == max(df_err.error)]
```

# 6.

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()

np.mean(cross_val_score(rf, X_train, y_train, scoring = 'neg_mean_absolute_error', cv=3))
```

# 7. 

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':range(10, 300, 10), 'criterion':('mse', 'mae'), 'max_features':('auto','sqrt','log2')}
gs = GridSearchCV(rf, parameters, scoring = 'neg_mean_absolute_error', cv=3)
gs.fit(X_train, y_train)
```

```python
gs.best_score_
```

```python
gs.best_estimator_
```

# 8.

```python
pred_lm = lm.predict(X_test)
pred_lml = lm_l.predict(X_test)
pred_rf = gs.best_estimator_.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test, pred_lm))
print(mean_absolute_error(y_test, pred_lml))
print(mean_absolute_error(y_test, pred_rf))
```

```python
mean_absolute_error(y_test, (pred_lm+pred_rf)/2)
```
